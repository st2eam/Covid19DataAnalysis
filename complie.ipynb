### 数据说明
数据更新至2022/05/17
- time_series_covid19_recovered_global.csv——全球covid19治愈数据
- time_series_covid19_deaths_global.csv——全球covid19死亡数据
- time_series_covid19_confirmed_global.csv——全球covid19确诊数据
- time_series_covid19_deaths_US.csv——美国covid19死亡数据
- time_series_covid19_confirmed_US.csv——美国covid19确诊数据
### 数据来源
COVID-19 Data Repository by the Center for Systems Science and Engineering (CSSE) at Johns Hopkins University

[https://github.com/CSSEGISandData/COVID-19](https://github.com/CSSEGISandData/COVID-19)

字段描述
  
Province_State:省份、州或依赖项名称。

Country_Region:国家、地区或主权名称。

Lat和Long_:仪表盘上的点位置。地图上显示的所有点(澳大利亚除外)都以地理中心为基础，在比省/州更细的空间尺度上不代表特定地址、建筑或任何位置。澳大利亚点位于每个州最大城市的中心。

Confirmed:计数包括已证实和可能(如有报告)。

Deaths:包括确诊和可能死亡(如有报告)。

Recovered:恢复病例是根据当地媒体报道以及可用的州和地方报道估算的，因此可能大大低于真实数字

```python
# 获取国家确诊人数最后两列
confirmed_global[confirmed_global['Province/State'].isna()==True][ ['Country/Region',confirmed_global.columns[-2], confirmed_global.columns[-1]] ]
# 计算每个国家当前新增确诊人数
df_countries = confirmed_global[confirmed_global['Province/State'].isna()==True][ ['Country/Region',confirmed_global.columns[-2], confirmed_global.columns[-1]] ]
df_countries['confirmed'] = df_countries[confirmed_global.columns[-1]] - df_countries[confirmed_global.columns[-2]]
df_countries.sort_values('confirmed', ascending=False)
# 按国家当前确诊人数前10名绘制柱状图
ax = df_countries.sort_values('confirmed', ascending=False)[:10].plot.bar(x='Country/Region', y='confirmed', 
title=confirmed_global.columns[-1]+ " 全球新冠确诊人数前十", figsize=(12, 6), rot=0)

# 标准柱状图
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() , p.get_height()+5 ))
```

## 获取中国区域数据
由于数据量较大，我们只分析中国最近两个月的数据  
将确诊数据计算为每日新增数据

```python
confirmed_China=confirmed_global[confirmed_global['Country/Region']=='China']
index = confirmed_China[confirmed_China["Province/State"] == "Unknown" ].index
confirmed_China.drop(index, axis = 0, inplace=True)
def handleData():
  for i in range(len(confirmed_China.columns)-1,4,-1):
    new = confirmed_China.iloc[:,i]-confirmed_China.iloc[:,i-1]
    confirmed_China.iloc[:,i] = new
handleData()
temp =  pd.concat([confirmed_China.iloc[:,:1],confirmed_China.iloc[:,-60:]],axis = 1)
new_confirmed_China = temp.iloc[:,1:].sum()
temp
```

```python
fig,axe=plt.subplots(figsize=(30,10),dpi=80)
axe.plot(new_confirmed_China.index,new_confirmed_China,'-o',color='c')
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False ## 设置正常显示符号
plt.title("全国近六十天新增确诊人数变化趋势图",fontsize=20)
plt.xticks(rotation=30,fontsize=14) # 旋转30度
plt.yticks(fontsize=14)
for a, b in zip(new_confirmed_China.index, new_confirmed_China):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=14)
plt.show()
```

```python
confirmed_China=confirmed_global[confirmed_global['Country/Region']=='China']
index = confirmed_China[confirmed_China["Province/State"] == "Unknown" ].index
confirmed_China.drop(index, axis = 0, inplace=True)
def handleData():
  for i in range(len(confirmed_China.columns)-1,4,-1):
    new = confirmed_China.iloc[:,i]-confirmed_China.iloc[:,i-1]
    confirmed_China.iloc[:,i] = new
handleData()
temp =  pd.concat([confirmed_China.iloc[:,:1],confirmed_China.iloc[:,-60:]],axis = 1)
temp["sum"] = temp.sum(axis=1)
temp.loc['93']=temp[temp["sum"]<1000].sum(axis=0)
temp["Province/State"]['93'] = "其他"
temp
```

```python
fig,axe = plt.subplots(figsize=(15,10),dpi=100)
data_new = temp[temp["sum"]>1000]
labels = data_new["Province/State"]
sizes = data_new["sum"]
colors = ['lightskyblue', 'gold', 'c','tomato', 'silver', '#f0944d']
axe.pie(sizes,textprops={'fontsize':12},labels=labels,colors=colors,autopct='%1.1f%%',startangle=180,pctdistance=0.8)
axe.axis('equal')
axe.set_title('各地区主要新增占比',fontsize=30)
plt.show()
```
